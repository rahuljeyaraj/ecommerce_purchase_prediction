```python
session_event_counts = pd.crosstab(df["user_session"], df["event_type"])
```

```python
session_event_counts.head()
```

```python
session_time = df.groupby("user_session")["event_time"].agg(["min", "max"])
session_time.head()
```

```python
session_time["min"] = pd.to_datetime(session_time["min"])
session_time["max"] = pd.to_datetime(session_time["max"])
session_time["session_duration_sec"] = (
    session_time["max"] - session_time["min"]
).dt.total_seconds()
```
